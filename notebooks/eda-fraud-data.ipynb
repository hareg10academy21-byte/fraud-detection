```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
fraud = pd.read_csv('../data/raw/Fraud_Data.csv')

fraud.head()
```

```python
fraud.shape
```

```python
fraud.info()
```

```python
fraud.describe()
```

```python
fraud.isnull().sum()
```

```python
missing = fraud.isnull().sum()

missing[missing > 0]
```

```python
fraud['age'].fillna(
    fraud['age'].median(),
    inplace=True
)
```

```python
fraud.duplicated().sum()
```

```python
fraud.drop_duplicates(inplace=True)
```

```python
fraud.duplicated().sum()
```

```python
fraud.info()
```

```python
fraud['signup_time'] = pd.to_datetime(
    fraud['signup_time']
)

fraud['purchase_time'] = pd.to_datetime(
    fraud['purchase_time']
)
```

```python
fraud.info()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    fraud['purchase_value'],
    kde=True
)

plt.title('Purchase Value Distribution')
plt.show()
```

Most purchases are concentrated at lower transaction values with a right-skewed distribution.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    fraud['age'],
    kde=True
)

plt.title('Age Distribution')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='browser',
    data=fraud
)

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='source',
    data=fraud
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='sex',
    data=fraud
)

plt.show()
```

```python
fraud['class'].value_counts()
```

```python
fraud['class'].value_counts(normalize=True)*100
```

```python
sns.countplot(
    x='class',
    data=fraud
)

plt.title('Fraud Distribution')
plt.show()
```

Fraudulent transactions represent only X% of the dataset, indicating severe class imbalance.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='class',
    y='purchase_value',
    data=fraud
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='class',
    y='age',
    data=fraud
)

plt.show()
```

```python
browser_fraud = pd.crosstab(
    fraud['browser'],
    fraud['class'],
    normalize='index'
)

browser_fraud
```

```python
source_fraud = pd.crosstab(
    fraud['source'],
    fraud['class'],
    normalize='index'
)

source_fraud
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    fraud.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```
